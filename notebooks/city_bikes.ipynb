```python
# Import python libraries

import pandas as pd
import requests
from dotenv import load_dotenv
import os
import numpy as np

```

```python
# Assess successful connection with CityBike's API
url = 'http://api.citybik.es/v2/networks'

response = requests.get(url)
```

```python
response.content
```

```python
response.json()

data = response.json()
```

```python
import pprint

pprint.pprint(data)
```

```python
# Extract list of stations from dictionary key 'networks'
data = data['networks']

# Count the number of citybike stations for data learning
len(data)
```

```python
# Format of each city's network
data[1]
```

```python
# Extracting Lisbon City details from API JSON file

city_list = []
network_id_list = []
latitude_list = []
longitude_list = []
country_list = []

for d in data:
    location = d.get('location','N/A')
    if location.get('city') == 'Lisbon' and location.get('country') == 'PT':
        network_id = d.get('id','N/A')
        city = location.get('city','N/A')
        latitude = location.get('latitude','N/A')
        longitude = location.get('longitude','N/A')
        country = location.get('country','N/A')


        network_id_list.append(network_id)
        city_list.append(city)
        latitude_list.append(latitude)
        longitude_list.append(longitude)
        country_list.append(country)

```

```python
# Creating DataFrame with Lisbon City details
bike_station_dic = {'city':city_list,
                    'network_id':network_id_list,
                    'latitude':latitude_list,
                    'longitude':longitude_list,
                    'country':country_list
                    }
                    
bike_station_df = pd.DataFrame(bike_station_dic)


Lisbon_id = bike_station_df['network_id'].iloc[0]

bike_station_df
```

```python
# Identifying Lisbon city's CityBike Network ID
print(f'Lisbon CityBike Network ID =',Lisbon_id)
```

```python
# Retrieving data from CtyBike's API filtered for stations from Lisbon city

url = 'http://api.citybik.es/v2/networks/gira?fields=stations'

response = requests.get(url)

response
```

```python
# prettyprint of Lisbon's bike stations
response.json()
lisbon_data = response.json()
pprint.pprint(data)
```

```python
# Reassigning variable into JSON file
lisbon_data = lisbon_data['network']['stations']
```

```python
# Identifying a list of dictionairies containing bike stations in chosen city
lisbon_data
```

Parse through the response to get the details you want for the bike stations in that city (latitude, longitude, number of bikes). 

```python
# Lists of bike_station data
latitude_list = []
longitude_list = []
timestamp_list = []
free_bikes_list = []
empty_slots_list = []
total_slots_list = []

# Iterate through all dictionaries of Lisbon's bike stations for chosen data
for ld in lisbon_data:
    latitude = ld.get('latitude','N/A')
    longitude = ld.get('longitude','N/A')
    timestamp = ld.get('timestamp','N/A')
    free_bikes = ld.get('free_bikes','N/A')
    empty_slots = ld.get('empty_slots','N/A')
    extra = ld.get('extra','N/A')
    total_slots = extra.get('slots','N/A')

    # Appending chosen data to lists
    latitude_list.append(latitude)
    longitude_list.append(longitude)
    timestamp_list.append(timestamp)
    free_bikes_list.append(free_bikes)
    empty_slots_list.append(empty_slots)
    total_slots_list.append(total_slots)
```

Put your parsed results into a DataFrame.

```python
# Creating a dictionary of lists
lb_bike_station_dic = {'latitude':latitude_list,
                    'longitude':longitude_list,
                    'timestamp':timestamp_list,
                    'free bikes':free_bikes_list,
                    'empty slots':empty_slots_list,
                    'total slots':total_slots_list
                    }
                    
# Creating a Pandas Dataframe for Lisbon Bike Stations
lb_bike_station_df = pd.DataFrame(lb_bike_station_dic)

lb_bike_station_df
```

```python
# Download Lisbon's Bike Station DataFrame for further analysis in other notebooks

lb_bike_station_df.to_csv('lisbon_bike_station_df.csv', index=False, header=True, sep=',')
```
